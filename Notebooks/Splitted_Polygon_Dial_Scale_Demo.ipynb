```python
start_inner_radius: float = 5
stop_inner_radius: float = 5
start_outer_radius: float = 6
stop_outer_radius: float = 8
start_angle_deg: float = -135
stop_angle_deg: float = 135
num_Major_Ticks: int = 7
face_counts: int = 30
distance: float = .2

polys = calc_splitted_poly(start_inner_radius = start_inner_radius,
                       end_inner_radius = stop_inner_radius,
                       start_outer_radius = start_outer_radius,
                       end_outer_radius = stop_outer_radius,
                       start_angle_deg = start_angle_deg,
                       end_angle_deg = stop_angle_deg,
                       segment_counts = num_Major_Ticks,
                       face_counts = face_counts,
                       distance = distance)

xy_lines=[]
for poly in polys:
    x_line, y_line = xy_array_to_xANDyArray(poly)
    
    xy_lines.append([x_line, y_line])
plot_multi_poly_x_y(xy_lines, False, filled=False, outline = True, title = "Splitted Scale Demo (outline)")
plot_multi_poly_x_y(xy_lines, False, filled=True, outline = False, title = "Splitted Scale Demo (filled)")
plot_multi_poly_x_y(xy_lines, False, filled=True, outline = True, title = "Splitted Scale Demo (filled & outline)")
```
